# What and why we are using Matplotlib

### What is it?
- This is a plotting library for Python programming and its numerical mathematics extension Numpy.
- John Hunter created it to try to replicate Matlab's plotting capabilities in Python.

### Why we using it?
- It have ability to play well with many operating system and graphic backends.
- Its also provide an object-oriented API for embedding plots into aplications using GUI toolkits like Tkinter, wxPython, Qt or GTK.

### Explore the official Matplotlib web page: https://matplotlib.org/

```python
# import library
import matplotlib.pyplot as plt 

# This line is only for jupyter notebook. This sets the backend
# of matlotlib to the 'inline' backend.
# Its mean that the output of plotting command is displayed
# 'inline' within frontend, directly below the code cell
%matplotlib inline
```

### Basic commands

- I recommend you to explore the document for details or you can use 'shift + tab' to check the document inline your code

```python
import numpy as np

# create a data to plot
x = np.linspace(0,5,5) # linspace use for create sequences
y = x ** 2
```

```python
x
```

```python
y
```

**plotting simple graph $y=x^2$**

```python
# After we have the data above
plt.plot(x,y) # use to plot the graph with given point (x,y)
plt.xlabel('X axis') # set title for x axis
plt.ylabel('Y axis') # set title for y axis
plt.title('Y = X^2') # set title for the graph
plt.show() # display the graph
```

- You can see that the graph it is not smooth because we do not have a lot of points to connect so we increase the number of point

```python
x = np.linspace(0,5,100) # increase to 100 points compare to 5 points before
y = x**2

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Y = X^2')
plt.show()
```

**Create multiple plot side by side**
- plt.subplot(#number_of_rows, #number_of_columns, #index)

```python
plt.subplot(1,2,1) # create position have 1 row, 2 columns, at index 1
plt.plot(x,x**2,'b--') # create the graph at the position above, check out document for more type of color
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = X^2')

plt.subplot(1,2,2) # at index 2
plt.plot(x,-x**2,'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = -X^2')
```

### Matplotlib Object Oriented
- The idead is we will create figure objects and then call methods or attributes from that objects.

```python
# Create Figure(Its absolutely empty)
fig = plt.figure()
```

```python
# Add set of axes to the figure we have just created
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range from 0 to 1)
```

```python
fig
```

```python
# we have this axes then we want to plot on this
axes.plot(x,x**2)

# Because the approach is OOP so we call the method set_
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title for this graph using OOP')
fig
```

- At the beginning you will struggle a little bit when doing this, but the advantage is that we can have full control where the figure are placed and we can easily add many of axes to the figure if you want.

**Add more than one axis**

```python
# first let create empty figure
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,0.8,0.8]) # using this as main axes
axes2 = fig1.add_axes([0.23,0.5,0.3,0.3]) # add inset axes

# plot the main axes1
axes1.plot(x,x**2,'b')
axes1.set_xlabel('X axes1')
axes1.set_ylabel('Y axes1')
axes1.set_title('Title for axes1')

# plot the inset axes2
axes2.plot(x,-x**2,'r')
axes2.set_xlabel('X axes2')
axes2.set_ylabel('Y axes2')
axes2.set_title('Title for axes2')
```
