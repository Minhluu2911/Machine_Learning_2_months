# What and why we are using Matplotlib

### What is it?
- This is a plotting library for Python programming and its numerical mathematics extension Numpy.
- John Hunter created it to try to replicate Matlab's plotting capabilities in Python.

### Why we using it?
- It have ability to play well with many operating system and graphic backends.
- Its also provide an object-oriented API for embedding plots into aplications using GUI toolkits like Tkinter, wxPython, Qt or GTK.

### Explore the official Matplotlib web page: https://matplotlib.org/

```python
# import library
import matplotlib.pyplot as plt 

# This line is only for jupyter notebook. This sets the backend
# of matlotlib to the 'inline' backend.
# Its mean that the output of plotting command is displayed
# 'inline' within frontend, directly below the code cell
%matplotlib inline
```

### Basic commands

- I recommend you to explore the document for details or you can use 'shift + tab' to check the document inline your code

```python
import numpy as np

# create a data to plot
x = np.linspace(0,5,5) # linspace use for create sequences
y = x ** 2
```

```python
x
```

```python
y
```

**plotting simple graph $y=x^2$**

```python
# After we have the data above
plt.plot(x,y) # use to plot the graph with given point (x,y)
plt.xlabel('X axis') # set title for x axis
plt.ylabel('Y axis') # set title for y axis
plt.title('Y = X^2') # set title for the graph
plt.show() # display the graph
```

- You can see that the graph it is not smooth because we do not have a lot of points to connect so we increase the number of point

```python
x = np.linspace(0,5,100) # increase to 100 points compare to 5 points before
y = x**2

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Y = X^2')
plt.show()
```

**Create multiple plot side by side**
- plt.subplot(#number_of_rows, #number_of_columns, #index)

```python
plt.subplot(1,2,1) # create position have 1 row, 2 columns, at index 1
plt.plot(x,x**2,'b--') # create the graph at the position above, check out document for more type of color
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = X^2')

plt.subplot(1,2,2) # at index 2
plt.plot(x,-x**2,'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Y = -X^2')
```

### Matplotlib Object Oriented
- The idead is we will create figure objects and then call methods or attributes from that objects.

```python
# Create Figure(Its absolutely empty)
fig = plt.figure()
```

```python
# Add set of axes to the figure we have just created
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range from 0 to 1)
```

```python
fig
```

```python
# we have this axes then we want to plot on this
axes.plot(x,x**2)

# Because the approach is OOP so we call the method set_
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title for this graph using OOP')
fig
```

- At the beginning you will struggle a little bit when doing this, but the advantage is that we can have full control where the figure are placed and we can easily add many of axes to the figure if you want.

**Add more than one axis**

```python
# first let create empty figure
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,0.8,0.8]) # using this as main axes
axes2 = fig1.add_axes([0.23,0.5,0.3,0.3]) # add inset axes

# plot the main axes1
axes1.plot(x,x**2,'b')
axes1.set_xlabel('X axes1')
axes1.set_ylabel('Y axes1')
axes1.set_title('Title for axes1')

# plot the inset axes2
axes2.plot(x,-x**2,'r')
axes2.set_xlabel('X axes2')
axes2.set_ylabel('Y axes2')
axes2.set_title('Title for axes2')
```

**subplots()** (note that it different with subplot)
- it creates a figure and a grid of subplots with a single call

```python
# a subplot without arguments return a Figure and a single Axes
fig, axes = plt.subplots()

# Now we use axes object to add graph
axes.plot(x,x**2,'r')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

**Figure size, aspect ratio and DPI**
- you can specific these thing when object is created
- `figsize` is a tuple of (width,height) of the figure in inches. The default size is (6.4,4.8) inches in matplotlib

- `dpi` is the dots-per-inch: determine how many pixels the figure comprises. The default is 100
    - px, py = w*dpi, h*dpi #
    - \# e.g
    - \# 6.4 inches * 100dpi = 640 pixels

```python
# let take example
fig = plt.figure(figsize=(8,4), dpi=100)
```

### Legends, labels and titles
- These thing is used to decorate a figure

```python
x1 = np.linspace(0,10,100)

fig1 = plt.figure(figsize=(6,4), dpi=100)

ax1 = fig1.add_axes([0.1,0.1,0.8,0.7])
```

**Figure titles**

```python
ax1.set_title('Title')

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
fig1.tight_layout()
fig1
```

**Legend**
- That is an area describing the elements of the graph.

```python
ax1.plot(x, x**2, label='$x^2$')
ax1.plot(x, x**3, label='$x^3$')

ax1.legend() # that used to call the legend to place it in the graph

fig1
```

- The legend function take an optional argument `loc` that can be used to specific where in the figure the legend is to be drawn.
- Checkout this [document](https://matplotlib.org/2.0.2/users/legend_guide.html) for more detail

```python
ax1.legend(loc=1) # upper right corner
fig1
```

```python
ax1.legend(loc=2) # upper left corner
fig1
```

```python
ax1.legend(loc=3) # lower left corner
fig1
```

```python
ax1.legend(loc=4) # lower right corner
fig1
```

```python
# Most common to choose
ax1.legend(loc=0) # let matplotlib choose optimal location
fig1
```

**Setting colors, linewidths, linetypes**
- `b` means blue, `g` means green, `r` means red


```python
x2 = np.linspace(0,5,10)

fig2, ax2 = plt.subplots()
ax2.plot(x2, x2**2, 'b.-') # blue line with dots
ax2.plot(x2, x2**3, 'g--') # green dashed line
```

- We can also using:
    - `color`: specific color by their name or RGB hex code and 
    - `alpha`: adjust tranparency of a graph

```python
x3 = np.linspace(0,5,10)

fig3, ax3 = plt.subplots()

ax3.plot(x3, x3, color="blue") # no transparant
ax3.plot(x3, x3+1, color="blue", alpha=0.5) # half-transparant

ax3.plot(x3, x3+2, color="#8B008B")        # RGB hex code
ax3.plot(x3, x3+3, color="#FF8C00")        # RGB hex code 
```

**Line and marker styles**
- `linewidth` or `lw`: to change the line width
- `linestyle` or `ls`: to change the linestyle

*linewidth*

```python
x = np.linspace(0,5,10)
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```

*linestyle*

```python

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
fig
```

*marker*

```python
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 8, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+9, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+10, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+11, color="blue", lw=3, ls='--', marker='1')
fig
```

*marker size and color*

```python
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
fig
```

**Control axis appearance**

```python
x = np.linspace(0,5,100)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3) 
axes[0].set_title("default axes ranges") 

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight') # automatic fit axes range
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60]) # set the range for y axes
axes[2].set_xlim([2, 5]) # set the range for y axes
axes[2].set_title("custom axes range");
```

### Special plot types

**Scatter plot**

```python
x = np.linspace(0,5,15)
y = x**2
plt.scatter(x,y)
```

**Histogram plot**

```python
from random import sample 
# sample return k length new list of elements chosen from the sequence.

x = sample(range(1,1000),100)
plt.hist(x)
```

**Box plot**

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);   
```

### Saving Figure
- Matplotlib can generate high-quality output with many format as PNG, JPG, EPS, SVG, PGF, PDF
- use the `savefig` method

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

x = np.linspace(0,5,20)

ax.plot(x,x**2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')

# saving image
fig.savefig('filename.png')
fig.savefig('filename.png', dpi=200)
fig.savefig('filename.pdf')
```
