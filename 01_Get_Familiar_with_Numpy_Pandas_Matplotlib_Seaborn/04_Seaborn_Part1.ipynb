```python
# we need to import its library
import seaborn as sns
%matplotlib inline
```

```python
# load built-in dataset of seaborn
tips = sns.load_dataset('tips')

# show the first 5 row of dataset
tips.head()
```

### [displot](https://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot)

- It is the combination of histogram with the seaborn kdeplot() and rugplot() function.
- It shows the distribution of a [univariate set of observation](https://en.wikipedia.org/wiki/Univariate_(statistics)#:~:text=Univariate%20is%20a%20term%20commonly,salaries%20of%20workers%20in%20industry.).

```python
sns.distplot(tips['total_bill'])
```

```python
# We can remove the kdeplot
sns.distplot(tips['total_bill'], kde=False)
```

```python
# Adjust the number of bins
sns.distplot(tips['total_bill'], bins=50)
```

```python
# Plot the distribution on the vertical axis and change the color
sns.distplot(tips['total_bill'], vertical=True, color="#87a832")
```

### [jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

- It used to explore relation ship between [bivariate data](https://en.wikipedia.org/wiki/Bivariate_data), as well as their distribution at the same time.

- There are several kind of plot to compare:
    - [scatter](https://chartio.com/learn/charts/what-is-a-scatter-plot/): use to observe relationship between 2 variable
    - [reg](https://seaborn.pydata.org/generated/seaborn.regplot.html): use to plot data and a linear regression model fit
    - [resid](https://www.statisticshowto.com/residual-plot/#:~:text=A%20residual%20plot%20is%20typically,that%20is%20non%2Dlinearly%20associated.): use to find problem with regression
    - [kde](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot): use for visualizing the probability density of a continuous variable
    - [hex](https://www.tutorialspoint.com/how-can-seaborn-library-be-used-to-display-a-hexbin-plot-in-python): when data is scattered unevently, it gets difficult to capture all the data point in a scatter plot then hexagonal binning comes into play.

```python
# scatter
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
# reg
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# resid
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
# kde
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# hex
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

### [pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

- It show pairwise relationship in a dataset ( just for the numerical columns)
- Support a color hue argument ( for categorical columns )

```python
tips.head()
```

```python
# numerical columns
sns.pairplot(tips)
```

```python
# adding hue for categorical columns
sns.pairplot(tips, hue='sex')
```

```python
# change markers for categorical
sns.pairplot(tips, hue='smoker', markers =['s','o'])
```

```python
# force marginal to histogram or any kind of plot
sns.pairplot(tips, hue='sex', diag_kind='hist')
```

```python
# Force both diagonal and off-diagonal plotting style
sns.pairplot(tips,kind='kde')
```

### [rugplot](https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot)

- It is a plot of data for a single quantitative variable along an axis
- For simple you can imagine it is analogous to a histogram with zero-width bins.

```python
sns.rugplot(tips['total_bill'])
```

### [kdeplot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot)

- kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
